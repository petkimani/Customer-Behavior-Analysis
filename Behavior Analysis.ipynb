```python
# check missing values
df.isnull().sum()
```

```python
# input missing values in "Review Rating" with median rating of each category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# change all column names into lower case and replace space with '_'
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
# rename 'purchase_amount_(usd)' to 'purchase_amount'
df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'}, inplace=True)
```

```python
df.columns
```

```python
# create a new column called age group then use qcut function to group the ages into 4 groups
labels = ['Young-Adult', 'Adult', 'Middle-Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days based on 'Frequency of Purchases' column
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Annually': 365
}


df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
# Confirm if contents between these two columns are the same
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# drop 'promo_code_used' column
df.drop('promo_code_used', axis=1, inplace=True)
```
